```python
extractor = GoProTelemetryExtractor(filepath)
extractor.open_source()
accl, accl_t = extractor.extract_data("ACCL")
gps5, gps5_t = extractor.extract_data("GPS5")
gps, gps_t = extractor.extract_data("GPS9")
gyro, gyro_t = extractor.extract_data("GYRO")
magn, magn_t = extractor.extract_data("MAGN")
face, face_t = extractor.extract_data("FACE")
facenm, facenm_t = extractor.extract_data("FACE")
corin, corin_t = extractor.extract_data("CORI")
iori, iori_t = extractor.extract_data("IORI")
grav, grav_t = extractor.extract_data("GRAV")
extractor.close_source()
```

```python
# Create subplot with 3 rows for X, Y, Z acceleration
fig = make_subplots(
    rows=3, cols=1,
    subplot_titles=('Z-Axis Acceleration', 'X-Axis Acceleration', 'Y-Axis Acceleration'),
    vertical_spacing=0.08,
    shared_xaxes=True
)

# Add X-axis acceleration (row 1)
fig.add_trace(
    go.Scatter(x=accl_t, y=accl[:, 0], 
               mode='lines', 
               name='Accel Z',
               line=dict(color='red', width=1)),
    row=1, col=1
)

# Add Y-axis acceleration (row 2)  
fig.add_trace(
    go.Scatter(x=accl_t, y=accl[:, 1], 
               mode='lines', 
               name='Accel X',
               line=dict(color='green', width=1)),
    row=2, col=1
)

# Add Z-axis acceleration (row 3)
fig.add_trace(
    go.Scatter(x=accl_t, y=accl[:, 2], 
               mode='lines', 
               name='Accel Y',
               line=dict(color='blue', width=1)),
    row=3, col=1
)

# Update layout
fig.update_layout(
    height=800,
    title_text="GoPro Accelerometer Data vs Time",
    showlegend=True
)

# Update x-axis labels
fig.update_xaxes(title_text="Time (seconds)", row=3, col=1)

# Update y-axis labels
fig.update_yaxes(title_text="Acceleration Z (m/s²)", row=1, col=1)
fig.update_yaxes(title_text="Acceleration X (m/s²)", row=2, col=1)
fig.update_yaxes(title_text="Acceleration Y (m/s²)", row=3, col=1)

fig.show()
```

```python
def arrays_to_datetime(days_since_2000, seconds_since_midnight):
    # Reference date: January 1, 2000
    reference_date = datetime(2000, 1, 1)
    
    # Convert to pandas datetime
    datetime_series = pd.to_datetime(reference_date) + \
                     pd.to_timedelta(days_since_2000, unit='D') + \
                     pd.to_timedelta(seconds_since_midnight, unit='s')
    
    return datetime_series
```

```python
gps_datetime = arrays_to_datetime(gps[:, 5], gps[:, 6])
```

```python
# days since 2000, secs since midnight (ms precision), DOP, fix (0, 2D or 3D)


fig = make_subplots(
    rows=5, cols=1,
    subplot_titles=('Days since 2000', 'Seconds since midnight', 'DOP', 'Fix', 'datetime'),
    vertical_spacing=0.08,
    shared_xaxes=True
)

# days since 2000
fig.add_trace(
    go.Scatter(x=gps_t, y=gps[:, 5], 
               mode='lines', 
               name='Days since 2000',
               line=dict(color='red', width=1)),
    row=1, col=1
)

# Add Y-axis acceleration (row 2)  
fig.add_trace(
    go.Scatter(x=gps_t, y=gps[:, 6], 
               mode='lines', 
               name='Seconds since midnight',
               line=dict(color='green', width=1)),
    row=2, col=1
)

# DOP
fig.add_trace(
    go.Scatter(x=gps_t, y=gps[:, 7], 
               mode='lines', 
               name='DOP',
               line=dict(color='blue', width=1)),
    row=3, col=1
)

# fix
fig.add_trace(
    go.Scatter(x=gps_t, y=gps[:, 8], 
               mode='lines', 
               name='Fix',
               line=dict(color='purple', width=1)),
    row=4, col=1
)

# datetime
fig.add_trace(
    go.Scatter(x=gps_t, y=gps_datetime, 
               mode='lines', 
               name='datetime',
               line=dict(color='orange', width=1)),
    row=5, col=1
)

# Update layout
fig.update_layout(
    height=800,
    title_text="GoPro GPS  Data vs Time",
    showlegend=True
)

# Update x-axis labels
fig.update_xaxes(title_text="Time (seconds)", row=3, col=1)

# Update y-axis labels
fig.update_yaxes(title_text="Days since 2000", row=1, col=1)
fig.update_yaxes(title_text="Seconds since midnight", row=2, col=1)
fig.update_yaxes(title_text="DOP", row=3, col=1)
fig.update_yaxes(title_text="Fix", row=4, col=1)
fig.update_yaxes(title_text="datetime", row=5, col=1)

fig.show()
```
